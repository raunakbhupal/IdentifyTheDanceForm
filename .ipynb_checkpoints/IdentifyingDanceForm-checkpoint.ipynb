```python
import os
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train=pd.read_csv('dataset/train.csv')
test=pd.read_csv('dataset/test.csv')
```

```python
train_dir=os.path.join('dataset/train/')
test_dir=os.path.join('dataset/test/')
```

```python
train_imgs=os.listdir(train_dir)
test_imgs=os.listdir(test_dir)
```

```python
train_imgs[:5]
```

```python
test_imgs[:5]
```

```python
#loading images of same size
#thus changing the size to 224*224
width=224
height=224
```

```python
def fn1(train,train_dir,train_imgs):
    train_data=[]
    train_label=[]
    for image in train_imgs:
        train_data.append(cv2.resize(cv2.imread(train_dir+image),(width,height),cv2.INTER_AREA))
        if image in list(train['Image']):
            train_label.append(train.loc[train['Image']==image,'target'].values[0])
    return train_data,train_label
```

```python
def fn2(test_dir,test_imgs):
    test_data=[]
    for image in test_imgs:
        test_data.append(cv2.resize(cv2.imread(test_dir+image),(width,height),cv2.INTER_AREA))
    return test_data
```

```python
train_data,train_label=fn1(train,train_dir,train_imgs)
```

```python
test_data=fn2(test_dir,test_imgs)
```

```python
#visualizing the images
plt.figure(figsize=(15,15))
for i in range(16):
    ax=plt.subplot(4,4,i+1)
    plt.imshow(train_data[i])
    plt.title(train_label[i])
    plt.axis('off')
```

```python
#to convert the string target into a numeric target
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
train_label=le.fit_transform(train_label)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(train_data,train_label,test_size=0.3,random_state=101)
```

```python
#some image augmentation to avoid over fitting
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagenerator = ImageDataGenerator(
    rescale=1. / 255,
    featurewise_center=False,  
    samplewise_center=False,  
    featurewise_std_normalization=False,  
    samplewise_std_normalization=False,  
    rotation_range=30,  
    zoom_range = 0.10,  
    width_shift_range=0.10,  
    height_shift_range=0.10,  
    horizontal_flip=True,  
    vertical_flip=False) 

val_datagenerator=ImageDataGenerator(rescale=1. / 255)

train_datagenerator.fit(X_train)
val_datagenerator.fit(X_test)
X_train=np.array(X_train)
X_test=np.array(X_test)
```

```python
X_train.shape
```

```python
#training through transfer learning

from tensorflow.keras.applications.vgg16 import VGG16
```

```python
vgg=VGG16(weights='imagenet',include_top=False,input_shape=(224,224,3),pooling='max')

vgg.summary()
```

```python
from tensorflow.keras.layers import Flatten,Dense,Dropout
from tensorflow.keras.models import Sequential
```

```python
vgg.trainable=False
model=Sequential()
model.add(vgg)
model.add(Dense(1024,activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(8,activation='softmax'))
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau
```

```python
reduce_lr=ReduceLROnPlateau(monitor='loss',factor=0.1,patience=2,verbose=1,min_lr=0.0001,cooldown=2)
callbacks=[reduce_lr]
```

```python
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
history=model.fit_generator(
    train_datagenerator.flow(X_train,to_categorical(y_train,8),batch_size=16),
    validation_data=val_datagenerator.flow(X_test,to_categorical(y_test,8),batch_size=16),
    shuffle=False,
    verbose=1,
    epochs=30,
    callbacks=callbacks
)
```

```python
acc=history.history['acc']
val_acc=history.history['val_acc']
loss=history.history['loss']
val_loss=history.history['val_loss']

epochs=range(len(acc))

plt.plot(epochs,acc,label='training')
plt.plot(epochs,val_acc,label='validation')
plt.legend()
plt.title('Training and validation accuracy')
```

```python
plt.plot(epochs,loss,label='training')
plt.plot(epochs,val_loss,label='validation')
plt.legend()
plt.title('Training and validation loss')
```

```python
predictions=model.predict_classes(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import f1_score
```

```python
f1=f1_score(y_test,predictions,average='weighted')
```

```python
score=100*f1
score
```

```python
final_df=pd.DataFrame(test_imgs,columns=['Image'])
```

```python
final_df.head()
```

```python
prediction_test=[]
for i in range(len(final_df)):
    image=final_df['Image'][i]
    resized_img=cv2.resize(cv2.imread(test_dir+image),(width,height),cv2.INTER_AREA)
    pred=model.predict_classes(resized_img.reshape(1,224,224,3))
    pred=le.inverse_transform(pred)
    prediction_test.append(pred[0])
```

```python
final_df['target']=prediction_test
```
